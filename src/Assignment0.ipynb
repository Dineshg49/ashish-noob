## Loading Images with Open-CV

```python
img1 = cv2.imread('img1.jpg')
print(img1)
```

```python
# Reading an image in OpenCV
img1 = cv2.imread('img1.jpg')

img2 = cv2.imread('img2.jpg')
img3 = cv2.imread('img3.jpg')
img4 = cv2.imread('img4.jpg')

#Saving Images
os.chdir('../images')
cv2.imwrite('img1.jpg',img1)
cv2.imwrite('img2.jpg',img2)
cv2.imwrite('img3.jpg',img3)
cv2.imwrite('img4.jpg',img4)

#
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)
```

## Displaying Images

```python


fig = plt.figure(figsize=(16,12))

# Plot the first image
ax = fig.add_subplot(221)
ax.imshow(img1)
ax.set_title('Image 1')
ax.axis("off")

# Plot the second image
ax = fig.add_subplot(222)
ax.imshow(img2)
ax.set_title('Image 2')
ax.axis("off")

# Plot the third image
ax = fig.add_subplot(223)
ax.imshow(img3)
ax.set_title('Image 2')
ax.axis("off")

# Plot the fourth image
ax = fig.add_subplot(224)
ax.imshow(img4)
ax.set_title('Image 2')
ax.axis("off")

plt.show()
```

## Converting To Grayscale Images and saving it to images folder

```python
r, g, b = img3[:,:,0], img3[:,:,1], img3[:,:,2]
gray_image = 0.2989 * r + 0.5870 * g + 0.1140 * b

os.chdir('../images')
cv2.imwrite('gray_img3.jpg',gray_image)

gray = cv2.imread('gray_img3.jpg')
plt.imshow(gray)
```

## Range of Values

```python
max_channels = np.amax([np.amax(img1[:,:,0]), np.amax(img1[:,:,1]), np.amax(img1[:,:,2])])
min_channels = np.amin([np.amin(img1[:,:,0]), np.amin(img1[:,:,1]), np.amin(img1[:,:,2])])

print("max channel value in img1 : ",max_channels)
print("min channel value in img1 : ",min_channels)
print()
max_channels = np.amax([np.amax(img2[:,:,0]), np.amax(img2[:,:,1]), np.amax(img2[:,:,2])])
min_channels = np.amin([np.amin(img2[:,:,0]), np.amin(img2[:,:,1]), np.amin(img2[:,:,2])])

print("max channel value in img2 : ",max_channels)
print("min channel value in img2 : ",min_channels)
print()

max_channels = np.amax([np.amax(img3[:,:,0]), np.amax(img3[:,:,1]), np.amax(img3[:,:,2])])
min_channels = np.amin([np.amin(img3[:,:,0]), np.amin(img3[:,:,1]), np.amin(img3[:,:,2])])

print("max channel value in img3 : ",max_channels)
print("min channel value in img3 : ",min_channels)
print()

max_channels = np.amax([np.amax(img4[:,:,0]), np.amax(img4[:,:,1]), np.amax(img4[:,:,2])])
min_channels = np.amin([np.amin(img4[:,:,0]), np.amin(img4[:,:,1]), np.amin(img4[:,:,2])])

print("max channel value in img4 : ",max_channels)
print("min channel value in img4 : ",min_channels)
print()
```

### It can be infered that the max value in each image is 255 and min value is 0.Thus the range of values is [0,255].
